## **1. Import Libraries**

````python
import torch
from transformers import pipeline
from IPython.display import Audio
````

## **2. Set up Prompt and Model**

````python
# Define a string variable containing the text to be converted to speech
text = "Python is a high-level, general-purpose programming language"

# Initialize the text-to-speech pipeline using the specified model and device
pipe = pipeline("text-to-speech", model="suno/bark-small", device="cuda")

# Use the pipeline to convert the text to speech and store the output
output = pipe(text)
````

````python
output
````

### **3. Create an Audio object to play the generated audio output**

````python
# 'output["audio"]' contains the audio data, and 'output["sampling_rate"]' specifies the playback rate
Audio(output["audio"], rate=output["sampling_rate"])
````
